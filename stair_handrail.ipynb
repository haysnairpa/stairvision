```python
!pip install ultralytics albumentations

from ultralytics import YOLO
from google.colab import drive

drive.mount('/content/drive')

model = YOLO('yolov8x-seg.pt')

model.train(
    data='/content/drive/MyDrive/stairvision/dataset/yolov5-seg-3/data.yaml',
    epochs=50,
    imgsz=640,
    batch=8,
    name='stairs_segmentation',
    augment=True,
    degrees=15,
    flipud=0.5,
    fliplr=0.5,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
)

results = model.predict('/content/drive/MyDrive/stairvision/dataset/yolov5-seg-3/train/images/frame_0005_jpg.rf.36c14ab12ca3c20500ec8a72a45fcad4.jpg', save=True, imgsz=640, conf=0.5)
results[0].show()
```
